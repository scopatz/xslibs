```python
import os

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
import numpy as np
import pandas as pd

relerror = lambda x, y: abs(x - y) / (abs(x + y)/2)

output_dir = os.path.expanduser("build-lwr1g/brightlite0")
outfile_path = os.path.join(output_dir, "fuel.txt")
klabel = 'Multiplication Factor'
bulabel = 'Burnup [MWd/kg]'
masslabel = 'Mass [g]'
```

## Xsgen Data

```python
#xdata = pd.read_csv(outfile_path, delim_whitespace=True, index_col=0).transpose()
xU235 = pd.read_csv(os.path.join(output_dir, '922350.txt'), delim_whitespace=True, index_col=0).transpose()
xU238 = pd.read_csv(os.path.join(output_dir, '922380.txt'), delim_whitespace=True, index_col=0).transpose()
xdata = xU235*0.05 + xU238*0.95
xdata.index = xdata.index.astype('f8')
plt.figure()
xdata.U235.plot(logy=True, figsize=(18, 9))
xdata.U238.plot(logy=True, figsize=(18, 9))
xdata.Pu239.plot(logy=True, figsize=(18, 9))
plt.legend(loc=1)
```

```python
plt.figure()
k = (xdata.NEUT_PROD / xdata.NEUT_DEST)
k.name = 'k'
k.plot(figsize=(18, 9))
plt.legend(loc=2)
BUd = xdata.BUd.cumsum()
BUd.name = 'BUd'
ax = BUd.plot(secondary_y=True)
ax.legend(loc=1)
```

## Bright-lite Comparison

```python
#xU235 = pd.read_csv(os.path.join(output_dir, '922350.txt'), delim_whitespace=True, index_col=0).transpose()
bU235 = pd.read_csv('/home/scopatz/.local/share/brightlite/extLWR/922350.txt', delim_whitespace=True, index_col=0).transpose()
bU238 = pd.read_csv('/home/scopatz/.local/share/brightlite/extLWR/922380.txt', delim_whitespace=True, index_col=0).transpose()
bdata = bU235*0.05 + bU238*0.95
bdata.U235 = bU235.U235*0.05
bdata.index = bdata.index.astype('f8')
```

```python
plt.figure()
xdata.U235.plot(logy=True, style='r-', figsize=(18, 9), label='U-235 (xsgen)')
xdata.U238.plot(logy=True, style='b-', figsize=(18, 9), label='U-238 (xsgen)')
xdata.Pu239.plot(logy=True, style='m-', figsize=(18, 9), label='Pu-239 (xsgen)')
bdata.U235.plot(logy=True, style='r--', figsize=(18, 9), label='U-235 (Bright-lite)')
bdata.U238.plot(logy=True, style='b--', figsize=(18, 9), label='U-238 (Bright-lite)')
bdata.PU239.plot(logy=True, style='m--', figsize=(18, 9), label='Pu-239 (Bright-lite)')
plt.legend(loc=0)
```

```python
plt.figure()
xk = (xdata.NEUT_PROD / xdata.NEUT_DEST)
xk.name = 'k'
xk.plot(style='b-', label='k (xsgen)', figsize=(18, 9))
bk = (bdata.NEUT_PROD / bdata.NEUT_DEST)
bk.name = 'k'
bk.plot(style='b--', label='k (Bright-lite)', figsize=(18, 9))
plt.legend(loc=2)
xBUd = xdata.BUd.cumsum()
xBUd.name = 'BUd'
ax = xBUd.plot(style='r-', label='BUd (xsgen)', secondary_y=True)
bBUd = bdata.BUd.cumsum()
bBUd.name = 'BUd'
ax = bBUd.plot(style='r--', label='BUd (Bright-lite)', secondary_y=True)
ax.legend(loc=1)
```

```python
plt.figure()
xk.index = xBUd.values
xk.plot(style='r-', label='xsgen', figsize=(18, 9))
bk.index = bBUd.values
bk.plot(style='b--', label='Bright-lite', figsize=(18, 9))
plt.xlabel(bulabel)
plt.ylabel(klabel)
plt.legend(loc=0)
```
