```python
with open(os.path.join(output_dir, 'xs.json'), 'r') as f:
    xsdata = json.load(f)
    
def filter_nuc_rx(nuc, rx):
    dat = []
    for t in xsdata:
        for row in t:
            if row[0] == nuc and row[1] == rx:
                dat.append(row[2])
    return dat
```

```python
fig = plt.figure(figsize=(18, 9))
times = xdata.index[:-1]
plt.semilogy(times, filter_nuc_rx('U235', 'fission'), 'r', label='$\sigma_f$, U-235', figure=fig)
plt.semilogy(times, filter_nuc_rx('U238', 'fission'), 'b', label='$\sigma_f$, U-238', figure=fig)
plt.semilogy(times, filter_nuc_rx('Pu239', 'fission'), 'm', label='$\sigma_f$, Pu-239', figure=fig)
plt.legend(loc=0)
```

```python
fig = plt.figure(figsize=(18, 9))
times = xdata.index[:-1]
plt.semilogy(times, filter_nuc_rx('U235', 'total'), 'r', label='$\sigma_t$, U-235', figure=fig)
plt.semilogy(times, filter_nuc_rx('U238', 'total'), 'b', label='$\sigma_t$, U-238', figure=fig)
plt.semilogy(times, filter_nuc_rx('Pu238', 'total'), 'm', label='$\sigma_t$, Pu-238', figure=fig)
plt.semilogy(times, filter_nuc_rx('H1', 'total'), 'g', label='$\sigma_t$, Pu-239', figure=fig)
plt.legend(loc=0)
```
